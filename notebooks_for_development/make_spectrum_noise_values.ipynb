```python
import glob
import os.path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
source_dir = ("/Users/nyumbani/Documents/git.repos/rrlyrae_metallicity/"+
              "rrlyrae_metallicity/src/synthetic_spectra_RW/rrmods_all/")
write_dir = ("/Users/nyumbani/Documents/git.repos/rrlyrae_metallicity/"+
             "rrlyrae_metallicity/src/synthetic_spectra_RW/junk_test_rrmods_all/")
```

```python
source_dir
```

```python
# accumulate list of filenames of normalized synthetic spectra
file_name_list = glob.glob(source_dir + "*smo") # (or search whatever other directory the *.c.dat files are in)

# for-loop to write out *.robolines and *.robospect files
for p in file_name_list: 
    print(p)
    df = pd.read_csv(p, delim_whitespace=True, names=["wavel","flux","error"])
    
    # make noise error equal to zero
    df["error"] = 1e-8
    
    df.to_csv(write_dir+os.path.basename(p), sep=' ', index=False, header=False)
```
