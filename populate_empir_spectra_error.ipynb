```python
# This takes the empirical spectra fluxes, and adds a column of the errors,
# which is just sqrt(flux)

# Created 2019 May 25 by E.S.
```

```python
import numpy as np
import pandas as pd
import glob
import os
```

```python
stem = "rrlyrae_metallicity/src/empirical_unnorm_spectra/backup/"
```

```python
file_list = glob.glob(stem + "*.dat")
```

```python
for i in range(0,len(file_list)):
    this_data = pd.read_csv(file_list[i], delim_whitespace=True, names = ["wavelength", "flux"])
    this_basename = os.path.basename(file_list[i])
    
    this_data["noise"] = np.sqrt(this_data["flux"])
    
    this_data.to_csv(stem + "../" + this_basename, sep = " ", header=False, index=False)
    print(this_basename)
```
